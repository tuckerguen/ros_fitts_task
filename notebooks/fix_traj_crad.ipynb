```python
import glob
text_files = glob.glob(r"C:\School\thesis\ros_fitts_task\trials\**\trial.pkl", recursive = True)
import sys
sys.path.append(r"C:\School\thesis\ros_fitts_task\src")
```

```python
import dill
trials = {}
for tf in text_files:
    with open(tf, "rb") as f:
        trials[tf] = dill.load(f)
```

```python
import re
def get_tsize(fn):
    reg = r".*\\.*_0\..*_(0.0.*?)\\.*"
    return re.match(reg, fn).groups()[0]
```

```python
for fn, trial in trials.items():
    for traj in trial:
        print(traj.crad, get_tsize(fn))
```
